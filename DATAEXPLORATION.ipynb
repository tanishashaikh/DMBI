```python

import pandas as pd
import statistics as s
import matplotlib.pyplot as plt
df = pd.read_csv('supermarket_sales - Sheet1.csv')
print(df)

df_tot = df.Total
print(df_tot)

df_mean = s.mean(df_tot)
print("Mean= ",df_mean)

df_med = s.median(df_tot)
print("Median= ",df_med)

df_mod = s.mode(df_tot)
print("Mode= ",df_mod)

df_min = min(df_tot)
print("Minimum value= ",df_min)

df_max = max(df_tot)
print("Maximum value= ",df_max)

fig=plt.figure(figsize=(10,8))
plt.boxplot(df_tot)
plt.show()
```
