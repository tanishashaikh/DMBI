```python
!pip install apyori
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

df = pd.read_csv('groceries - groceries.csv')

# print(df)

# print(np.shape(df))

list1=[]

for i in range(0, 9):
    list1.append([str(df.values[i,j]) for j in range(0,33)])
    
rule= apriori(transactions=list1, min_suport=0.003, min_confidence=0.2, min_lift=3, min_len=2, max_len=2)


result= list(rule)
# print(result)

for item in result:
    pair= item[0]
    items= [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    print("support" + str(item[0]))
    print("Confidence" + str(item[2][0][2]))
    print("support" + str(item[2][0][3]))
    print("=====================================")
```
